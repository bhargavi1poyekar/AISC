###Artificial Neural Network:

### Boolean Functions using M-P Neuron

1. **AND GATE:**

For AND function, output is high only if both inputs are high.
Assume weights as w1=w2=1

Net input calculation: ( y = x1w1 + x2w2 )

(0,0) y = 0x1 + 0x1 = 0

(0,1) y = 0x1 + 1x1 = 1

(1,0) y = 1x1 + 0x1 = 1

(1,1) y = 1x1 + 1x1 = 2

For both high inputs, net input =2.

Hence, threshold is equal to 2. If net input is greater than or equal to 2, the output is high.

output:

f(y)=1   if y>=2    
f(y)=0  if y<2

2. **AND NOT GATE:**

For AND NOT function, output is high only if first input is high and second input is low.
Assume one weight as excitatory and the other as inhibitory (w1=1, w2=-1)

Net input calculation: ( y = x1w1 + x2w2 )

(0,0) y = 0x1 + 0x-1 = 0

(0,1) y = 0x1 + 1x-1 = -1

(1,0) y = 1x1 + 0x-1 = 1

(1,1) y = 1x1 + 1x-1 = 0

For first high second low input, net input is 1

Hence, threshold is equal to 1. If net input is greater than or equal to 1, the output is high.

output:

f(y)=1   if y>=1

f(y)=0  if y<1

3. **XOR GATE:**

**First function:** (z1=x1.not(x2)) 


For first function, output is high only if first input is high and second input is low.
Assume one weight as excitatory and the other as inhibitory (w11=1, w21=-1)

Net input calculation: ( z1 = x1w11 + x2w21 )

(0,0) z1 = 0x1 + 0x-1 = 0

(0,1) z1 = 0x1 + 1x-1 = -1

(1,0) z1 = 1x1 + 0x-1 = 1

(1,1) z1 = 1x1 + 1x-1 = 0

For first high second low input, net input is 1

Hence, threshold is equal to 1. 

**Second function:** (z2=not(x1).x2) 


For second function, output is high only if first input is low and second input is high.
Assume one weight as inhibitory and the other as excitatory (w12=-1, w22=1)

Net input calculation: ( z2 = x1w12 + x2w22 )

(0,0) z1 = 0x-1 + 0x1 = 0

(0,1) z1 = 0x-1 + 1x1 = 1

(1,0) z1 = 1x-1 + 0x1 = -1

(1,1) z1 = 1x-1 + 1x1 = 0

For first low and second high input, net input is 1

Hence, threshold is equal to 1. 

**Third function:** (y=z1 or z2) 


For third function, output is high only if any one (not both) of the inputs is high.
Assume weights as v1=1, v2=1

Net input calculation: ( y = z1v1 + z2v2 )

(0,0) z1 = 0x1 + 0x1 = 0

(0,1) z1 = 0x1 + 1x1 = 1

(1,0) z1 = 1x1 + 0x1 = 1

(1,1) z1 = 0x1 + 0x1 = 0

For the required high output, net input is 1

Hence, threshold is equal to 1. 

output:

f(y)=1   if y>=1

f(y)=0  if y<1



4.  **OR GATE:**

For OR function, output is high if any one of the input is high.
Assume weights as w1=1, w2=1

Net input calculation: ( y = x1w1 + x2w2 )

(0,0) y = 0x1 + 0x1 = 0

(0,1) y = 0x1 + 1x1 = 1

(1,0) y = 1x1 + 0x1 = 1

(1,1) y = 1x1 + 1x1 = 2

For any one input high, minimum net input value is 1.

Hence, threshold is equal to 1. If net input is greater than or equal to 1, the output is high.

output:

f(y)=1   if y>=1

f(y)=0  if y<1



## CODE:

###AND GATE

```python
def AND(x1, x2):
  w1=w2=1                 # weights assigned
  y=x1*w1+x2*w2      # net input calculation
  if(y>=2):                 # threshold value = 2 
    return 1               # output high
  return 0                # output low
```

```python
x1,x2=map(int, input("Enter the input: ").split())   # taking inputs
op=AND(x1,x2)                                                     # calling AND function 
print("The output is : " + str(op))                         # printing output
```

### AND NOT GATE

```python
def AND_NOT(x1, x2):
  # weights assigned
  w1=1                     
  w2=-1
  y=x1*w1+x2*w2    # net input calculation
  if(y>=1):                # threshold value = 1
    return 1              # output high
  return 0               # output low
```

```python
x1,x2=map(int, input("Enter the input: ").split())  # taking inputs
op=AND_NOT(x1,x2)                                          # calling AND function 
print("The output is : " + str(op))                        # printing output
```

### XOR GATE

```python
def XOR(x1, x2):
  # first function (z1=x1. not(x2))
  w11=1
  w21=-1
  z1=x1*w11+x2*w21
  if z1>=1: #threshold = 1
    z1=1
  else:
    z1=0
  # second function (z1=not(x1). x2)
  w12=-1
  w22=1
  z2=x1*w12+x2*w22
  if z2>=1: #threshold = 1
    z2=1
  else:
    z2=0
  # third function (y=z1 or z2)
  v1=v2=1
  y=z1*v1+z2*v2
  if y>=1: #threshold = 1
    return 1
  return 0
```

```python
x1,x2=map(int, input("Enter the input: ").split())  # taking inputs
op=XOR(x1,x2)                                                     # calling AND function
print("The output is : " + str(op))                         # printing output
```

### OR GATE

```python
def OR(x1, x2):
  w1=w2=1             # weights assigned 
  y=x1*w1+x2*w2  # net input calculation
  if(y>=1):              # threshold value = 1
    return 1           # output high
  return 0            # output low
```

```python
x1,x2=map(int, input("Enter the input: ").split())  # taking inputs
op=OR(x1,x2)                                                       # calling AND function
print("The output is : " + str(op))                        # printing output
```
